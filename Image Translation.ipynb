# **Image Translation**

Write a Python program using OpenCV to perform image translation on the **"Given image"**.

The program should:

1. Import `cv2`, `numpy`, and `matplotlib.pyplot`.
2. Read the image **"given** in color mode.
3. Create a **translation matrix** that moves the image  
   **50 pixels to the right** and **80 pixels down**.
4. Apply the translation
5. Display the following using a **1 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Translated Image**


```python
# Import 

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image using OpenCV
img = cv2.imread('baseball.jpg')

```

```python
# Convert BGR to RGB for Matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Get image dimensions
rows, cols = img.shape[:2]

# Translation matrix
# Move 50 pixels right and 80 pixels down
T = np.float32([[1, 0, 50],
                [0, 1, 80]])

# Apply translation
translated = cv2.warpAffine(img_rgb, T, (cols, rows))
```

```python
# Display the image using Matplotlib

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(translated)
plt.title('Translated Image')
plt.axis('off')

plt.show()
```

# **Image Shearing**

Write a Python program using OpenCV to perform image shearing on the **given image**.

The program should:

1. Create a **shearing transformation matrix** to apply:
   - **Horizontal shearing**  
   - **Vertical shearing**
2. Apply the shearing transformations.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Sheared Image**  
   - **Vertically Sheared Image**


```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read image
img = cv2.imread('baseball.jpg')
```

```python
# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

```

```python
# Get dimensions
rows, cols = img_rgb.shape[:2]
```

```python
# Horizontal shearing matrix
H_shear = np.float32([[1, 0.5, 0],
                      [0, 1, 0]])
```

```python
# Vertical shearing matrix
V_shear = np.float32([[1, 0, 0],
                      [0.5, 1, 0]])
```

```python
# Apply shearing
horizontal_shear = cv2.warpAffine(
    img_rgb,
    H_shear,
    (int(cols + rows * 0.5), rows)
)

vertical_shear = cv2.warpAffine(
    img_rgb,
    V_shear,
    (cols, int(rows + cols * 0.5))
)
```

```python
# Display images
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(horizontal_shear)
plt.title('Horizontally Sheared')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(vertical_shear)
plt.title('Vertically Sheared')
plt.axis('off')

plt.show()
```

# **Image Reflection**

Write a Python program using OpenCV to perform image reflection on the **given image**.

The program should:

1. Generate the following reflected versions of the image:
   - **Horizontal Reflection** (flip left ↔ right)  
   - **Vertical Reflection** (flip top ↔ bottom)  
   - **Both-axis Reflection**  
2. Use `cv2.flip` with appropriate flip codes.
3. Display the following using a **2 × 2 layout** in Matplotlib:  
   - **Original Image**  
   - **Horizontally Reflected Image**  
   - **Vertically Reflected Image**  
   - **Both-axis Reflected Image**


```python
# Import libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread('baseball.jpg')
```

```python
# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

```

```python
# Horizontal Reflection (left ↔ right)
horizontal_flip = cv2.flip(img_rgb, 1)

```

```python
# Vertical Reflection (top ↔ bottom)
vertical_flip = cv2.flip(img_rgb, 0)

```

```python
# Both-axis Reflection
both_flip = cv2.flip(img_rgb, -1)

```

```python
# Display images
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(horizontal_flip)
plt.title('Horizontally Reflected')
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(vertical_flip)
plt.title('Vertically Reflected')
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(both_flip)
plt.title('Both-axis Reflected')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# **Image Rotation**

Write a Python program using OpenCV to perform **image rotation** on the **given image**.

The program should:

1. Rotate the image by:  
   - **45 degrees**  
   - **90 degrees**
2. Use `cv2.getRotationMatrix2D` and `cv2.warpAffine`.
3. Display the following using a **1 × 3 layout** in Matplotlib:  
   - **Original Image**  
   - **45° Rotated Image**  
   - **90° Rotated Image**


```python
# Import libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# Read the image
img = cv2.imread('baseball.jpg')

```

```python
# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Get image dimensions
rows, cols = img_rgb.shape[:2]
```

```python
# Find center of image
center = (cols // 2, rows // 2)
```

```python
# Rotation matrix for 45 degrees
M45 = cv2.getRotationMatrix2D(center, 45, 1)

```

```python
# Rotation matrix for 90 degrees
M90 = cv2.getRotationMatrix2D(center, 90, 1)

```

```python
# Apply rotation
rotated_45 = cv2.warpAffine(img_rgb, M45, (cols, rows))

rotated_90 = cv2.warpAffine(img_rgb, M90, (cols, rows))
```

```python
# Display images
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(rotated_45)
plt.title('45° Rotated Image')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(rotated_90)
plt.title('90° Rotated Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
